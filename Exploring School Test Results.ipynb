```python
import pandas as pd
file = pd.read_csv("schools.csv", encoding="latin1")
file=pd.DataFrame(file)
threshold = 0.80*800
filtered_schools = file[file["average_math"] >= threshold]
best_math_schools = filtered_schools[["school_name", "average_math"]]
best_math_school=filtered_schools.sort_values(by="average_math",ascending=False)
file["total_SAT"] = file["average_math"] + file["average_reading"] + file["average_writing"]
top_10_schools = file.sort_values(by=["total_SAT"], ascending=False).head(10)
borough_stats = file.groupby('borough')['total_SAT'].agg(['mean', 'std', 'count'])
max_std_row = borough_stats.loc[borough_stats['std'].idxmax()]
# Create a DataFrame to store the results
largest_std_dev = pd.DataFrame({
    'borough': [max_std_row.name],
    'num_schools': [max_std_row['count']],
    'average_SAT': [max_std_row['mean']],
    'std_SAT': [max_std_row['std']]
})

# Print the results
best_math_school=best_math_school.round(2)
top_10_schools=top_10_schools.round(2)
largest_std_dev=largest_std_dev.round(2)

print("Best Math Schools:")
print(best_math_schools)
print("\nTop 10 Schools by Total SAT:")
print(top_10_schools)
print("\nBorough with Largest Standard Deviation in Total SAT Scores:")
print(largest_std_dev)
```
