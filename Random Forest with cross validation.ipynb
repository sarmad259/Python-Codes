```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('/content/drive/Othercomputers/My Computer/Machine Learning/Datasets/heart.csv')
```

```python
data.sample(5)
```

```python
#Get Target data
y = data['target']

#Load X Variables into a Pandas Dataframe with columns
X = data.drop(['target'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
DT=DecisionTreeClassifier( )
RF=RandomForestClassifier()
DT.fit(X_train,y_train)
RF.fit(X_train,y_train)
```

```python
print(' Decision Tree Accuracy: {:.2f}'.format(DT.score(X_test, y_test)))
print(' Random Forest Accuracy: {:.2f}'.format(RF.score(X_test, y_test)))
```

Random Forest hyperparameters tuning using GridSearchCV


```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 100, num = 10)]
# Number of features to consider at every split
max_features = ['log2', 'sqrt']
# Maximum number of levels in tree
max_depth = [None,2,3,4,5]
#Number of samples to draw from X to train each base estimator
max_samples= [None,0.5,0.6,0.7]
```

```python
# Create the param grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'max_samples': max_samples}
print(param_grid)
```

```python
rf_Model = RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
rf_Grid = GridSearchCV(estimator = rf_Model, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)
```

```python
rf_Grid.fit(X_train, y_train)
```

```python
rf_Grid.best_params_
```

```python
print (f'Train Accuracy - : {rf_Grid.score(X_train,y_train):.3f}')
print (f'Test Accuracy - : {rf_Grid.score(X_test,y_test):.3f}')
```
