```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Extract the price data
price_data = laptop_data_mm[['Price']].to_numpy()

# Remove nan values from price_data
price_data = price_data[~np.isnan(price_data)]

# Check if there's data left after removing nan values
if price_data.shape[0] > 0:
    # Calculate Q1, Q3, and IQR
    Q1 = np.percentile(price_data, 25, axis=0)
    Q3 = np.percentile(price_data, 75, axis=0)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out outliers based on IQR
    mask = (price_data >= lower_bound) & (price_data <= upper_bound)
    price_data_no_outliers = price_data[mask]

    # Ensure there's data to scale
    if price_data_no_outliers.shape[0] > 0:
        min_max_scaler = MinMaxScaler(feature_range=(-1, 1))

        # Reshape price_data_no_outliers to a 2D array
        price_data_no_outliers = price_data_no_outliers.reshape(-1, 1)

        laptop_scaled = min_max_scaler.fit_transform(price_data_no_outliers)
        print("Q1:", Q1)
        print("Q3:", Q3)
        print("IQR:", IQR)
        print(laptop_scaled)
    else:
        print("No data left after outlier removal.")
else:
    print("No data left after removing nan values.")
```
