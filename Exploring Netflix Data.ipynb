```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

netflix_df = pd.read_csv("netflix_data.csv")

netflix_subset = netflix_df[netflix_df["type"] == "Movie"]

netflix_movies = netflix_subset[["title", "release_year", "duration"]]

short_movies = netflix_movies[netflix_movies.duration < 60]

X = short_movies[['release_year', 'duration']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
short_movies['cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
plt.scatter(short_movies['release_year'], short_movies['duration'], c=short_movies['cluster'], cmap='viridis')
plt.title('K-means Clustering of Short Movies')
plt.xlabel('Release Year')
plt.ylabel('Duration (min)')
plt.colorbar(label='Cluster')
plt.show()
My_Answer="No"
print("Are Movies Getting Shorter?")
print(f"My Answer is {My_Answer}")
```
